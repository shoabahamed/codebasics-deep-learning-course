<p>Divine flowers is a flower shop that sells different type of flowers. Below is the table showing how many flowers of each type they sold in different months. Also given are the prices of one flower each. Using this find out their total sales in every month.</p>



```python
units = np.array([[50, 60, 25], [10, 13, 5], [40, 70, 52]])
units
```

```python
price_per_unit = np.array([[20], [30], [15]])
price_per_unit
```

```python
# total sales in every month
```

```python
flower_income = np.dot(price_per_unit.reshape(-1, ), units)
flower_income
```
